```python
# manipulacja danymi
import numpy as np
import pandas as pd

# wizualizacja danych
import matplotlib.pyplot as plt
import seaborn as sns

# statystyczna analiza danych
from scipy import stats
```

```python
norm_rvs_100el =stats.norm.rvs(3, 1, 100)
norm_rvs_10000el =stats.norm.rvs(3, 1, 10000)
x_100 = np.linspace(-10, 10, 100)
pdf_100el= stats.norm.pdf(x_100, 3,1)
x_10000 = np.linspace(-10, 10, 10000)
pdf_10000el = stats.norm.pdf(x_10000, 3,1)
```

```python
fig,ax = plt.subplots(1, 1)
plt.xlim([-1, 7])
ax.hist(norm_rvs_100el,density=True,bins='auto',label='rvs 100el')
ax.plot(x_100, pdf_100el, label='pdf 100el')
ax.legend(loc='best', frameon=False)
```

```python
fig,ax = plt.subplots(1, 1)
plt.xlim([-1, 7])
ax.hist(norm_rvs_10000el,density=True,bins='auto',label='rvs 10000el')
ax.plot(x_10000, pdf_10000el, label='pdf 10000el')
ax.legend(loc='best', frameon=False)
```

rednia

```python
mean_100el =np.mean(norm_rvs_100el).round(3)
mean_10000el = np.mean(norm_rvs_10000el).round(3)
print('rednia dla 100el:',mean_100el)
print('rednia dla 10000el:',mean_10000el)
```

Moda.
W przypadku obydwu zbiorw szansa wygenerowania dwch takich samych liczb jest znikoma i dlatego te zrezygnowano z szukania mody.

Mediana

```python
median_100el = np.median(norm_rvs_100el).round(3)
median_10000el = np.median(norm_rvs_10000el).round(3)
print('Mediana dla 100el:',median_100el)
print('Mediana dla 10000el:',median_10000el)
```

Kwartyle

```python
q0_100 = np.quantile(norm_rvs_100el, 0.0).round(3)
q1_100 = np.quantile(norm_rvs_100el, 0.25).round(3)
q3_100 = np.quantile(norm_rvs_100el, 0.75).round(3)
q4_100 = np.quantile(norm_rvs_100el, 1.0).round(3)
print('Kwartyle dla 100el')
print(f'Q0: {q0_100}')
print(f'Q1: {q1_100}')
print(f'Q2: {median_100el}')
print(f'Q3: {q3_100}')
print(f'Q4: {q4_100}')
```

```python
q0_10000 = np.quantile(norm_rvs_10000el, 0.0).round(3)
q1_10000 = np.quantile(norm_rvs_10000el, 0.25).round(3)
q3_10000 = np.quantile(norm_rvs_10000el, 0.75).round(3)
q4_10000 = np.quantile(norm_rvs_10000el, 1.0).round(3)
print('Kwartyle dla 10000el')
print(f'Q0: {q0_10000}')
print(f'Q1: {q1_10000}')
print(f'Q2: {median_10000el}')
print(f'Q3: {q3_10000}')
print(f'Q4: {q4_10000}')
```

```python
plt.hist(norm_rvs_100el, density=True,bins='auto')
plt.axvline(x = mean_100el, color='red', label = 'rednia')
plt.axvline(x = median_100el, color='green', label = 'Mediana')
plt.axvline(x = q0_100, color='black', label = 'Min')
plt.axvline(x = q1_100, color='black', label = 'Q1')
plt.axvline(x = q3_100, color='black', label = 'Q3')
plt.axvline(x = q4_100, color='black', label = 'Max')
plt.legend(loc='best')
plt.show()
```

```python
plt.hist(norm_rvs_10000el, density=True,bins='auto')
plt.axvline(x = mean_10000el, color='red', label = 'rednia')
plt.axvline(x = median_10000el, color='green', label = 'Mediana')
plt.axvline(x = q0_10000, color='black', label = 'Min')
plt.axvline(x = q1_10000, color='black', label = 'Q1')
plt.axvline(x = q3_10000, color='black', label = 'Q3')
plt.axvline(x = q4_10000, color='black', label = 'Max')
plt.legend(loc='best')
plt.show()
```

Zakres

```python
range_100el = max(norm_rvs_100el) - min(norm_rvs_100el)
range_10000el = max(norm_rvs_10000el) - min(norm_rvs_10000el)
print('zakres dla 100el:',range_100el)
print('zakres dla 10000el:',range_10000el)
```

Rozstp midzykwartylowy - IQR

```python
iqr_100 = q3_100 - q1_100
iqr_10000 = q3_10000 - q1_10000
print('IQR dla 100 el:',round(iqr_100,3))
print('IQR dla 10000 el:',round(iqr_10000,3))
```

```python
plt.figure(figsize=(10,5))
plt.boxplot(norm_rvs_100el,vert=False)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.boxplot(norm_rvs_10000el,vert=False)
plt.show()
```

Wariancja

```python
variance_100el = np.var(norm_rvs_100el, ddof=1).round(3)
variance_10000el = np.var(norm_rvs_10000el, ddof=1).round(3)
print('wariacja dla 100el',variance_100el)
print('wariacja dla 10000el',variance_10000el)
```

```python
standard_100el = np.std(norm_rvs_100el, ddof=1).round(3)
standard_10000el = np.std(norm_rvs_10000el, ddof=1).round(3)
print('Odchylenie standardowe dla 100el',standard_100el)
print('Odchylenie standardowe dla 10000el',standard_10000el)
```

```python
list_title=['Liczba elementw zbioru','rednia','Mediana','Q0','Q1','Q2','Q3','Q4','Zakresy','IQR','Wariancje','Odchylenie standardowe']
list_100=[100,mean_100el,median_100el,q0_100,q1_100,median_100el,q3_100,q4_100,range_100el,iqr_100,variance_100el,standard_100el]
list_10000=[10000,mean_10000el,median_10000el,q0_10000,q1_10000,median_10000el,q3_10000,q4_10000,range_10000el,iqr_10000,variance_10000el,standard_10000el]
df= pd.DataFrame([list_100,list_10000])
df.columns=list_title
df.set_index('Liczba elementw zbioru')
```

Biorc pod uwag powysze dane, zbir 10000el charakteryzuj si wartociami redniej i mediany bliszymi wartoci zadanej 3. Wariancja i odchylenie standardowe dla tego zbioru rwnie jest blisze 1. IQR dla zbioru 10000 elementowego jest najbliszy wartoi 1.34896 wciewej dla rozkadu normalnego. 
Z powyszego wynika, i wikszy zbir daje wyniki najbardziej blione do tych charakterystycznych dla rozkadu normalnego.
